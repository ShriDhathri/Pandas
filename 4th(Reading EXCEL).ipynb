```python
import pandas as pd
def convert_people_cell(cell):
    if cell=="n.a.":
        return "Sam Waltson"
    return cell

def convert_eps_cell(cell):
    if cell=="not available":
        return None  #None will give NaN
    return cell
df=pd.read_excel(r"C:\Users\a\Documents\Python Scripts\stock.xlsx", converters={
    'people': convert_people_cell,
    'eps':convert_eps_cell
})
df
```

```python
df.to_excel("new.xlsx",sheet_name="Stocks", startrow=2, startcol=5, index=False)
```
